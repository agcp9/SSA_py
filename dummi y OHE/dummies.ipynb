```python
import pandas as pd

# Crear un DataFrame
df = pd.DataFrame({
    'Color': ['Rojo', 'Verde', 'Azul', 'Verde', 'Rojo', 'Azul']
})
df
```

```python
# Convertir la variable categ√≥rica en variables ficticias
dummy_variables = pd.get_dummies(df['Color'], drop_first=True)

# Mostrar el resultado
print(dummy_variables)
```

```python
# Aplicar One-Hot Encoding
one_hot_encoded = pd.get_dummies(df['Color'])

# Mostrar el resultado
print(one_hot_encoded)
```
