```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
```

```python
# Crear datos de ejemplo
np.random.seed(42)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + 1.5 * X**2 + np.random.randn(100, 1)
```

```python
# Visualizar datos de ejemplo
plt.scatter(X, y, label='Datos de ejemplo')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
# Aplicar regresión polinomial de segundo grado
poly_features = PolynomialFeatures(degree=2, include_bias=False)
X_poly = poly_features.fit_transform(X)
```

```python
lin_reg = LinearRegression()
lin_reg.fit(X_poly, y)
```

```python
# Visualizar la regresión polinomial
X_new = np.linspace(0, 2, 100).reshape(100, 1)
X_new_poly = poly_features.transform(X_new)
y_new = lin_reg.predict(X_new_poly)

plt.scatter(X, y, label='Datos de ejemplo')
plt.plot(X_new, y_new, 'r-', label='Regresión Polinomial (grado 2)')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```
