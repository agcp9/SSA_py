## ¬øC√≥mo funciona los arboles de clasificaci√≥n en Python?


En Python, los √°rboles de clasificaci√≥n se pueden implementar f√°cilmente utilizando bibliotecas de aprendizaje autom√°tico como Scikit-learn. Aqu√≠ te explico los pasos b√°sicos para crear un √°rbol de clasificaci√≥n en Python usando esta biblioteca:

- Preparar los datos: Antes de entrenar un modelo, necesitas preparar tus datos. Esto incluye dividir los datos en un conjunto de entrenamiento y un conjunto de prueba, y posiblemente normalizar o estandarizar los datos.

- Importar las bibliotecas necesarias: Primero, necesitas importar DecisionTreeClassifier de sklearn.tree y otras bibliotecas necesarias para el manejo de datos y la evaluaci√≥n del modelo.

- Cargar y preparar el conjunto de datos: Puedes utilizar un conjunto de datos existente como el Iris dataset o cargar uno propio. Luego, dividir√°s los datos en caracter√≠sticas (X) y la etiqueta objetivo (y), y despu√©s en conjuntos de entrenamiento y prueba.

- Entrenar el modelo: Crear√°s una instancia de DecisionTreeClassifier y utilizar√°s el m√©todo fit para entrenar el modelo con los datos de entrenamiento.

- Hacer predicciones: Una vez que el modelo est√° entrenado, puedes utilizarlo para hacer predicciones con nuevos datos o con el conjunto de prueba.

- Evaluar el modelo: Puedes evaluar el rendimiento del modelo utilizando m√©tricas como la precisi√≥n, que compara las etiquetas predichas con las reales.

- Visualizar el √°rbol: Scikit-learn tambi√©n ofrece herramientas para visualizar el √°rbol de decisi√≥n, lo que puede ayudarte a entender c√≥mo el modelo toma decisiones.

# Cosas a tener en cuenta

Es crucial recordar que, aunque los √°rboles de regresi√≥n son herramientas poderosas, tambi√©n tienen limitaciones, como la tendencia al sobreajuste, especialmente con √°rboles muy profundos. Por lo tanto, es com√∫n aplicar t√©cnicas como la poda del √°rbol o usar ensamblajes de √°rboles (por ejemplo, bosques aleatorios o gradient boosting) para mejorar la generalizaci√≥n del modelo.


```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import pandas as pd
```

```python
# Cargando un conjunto de datos de ejemplo
from sklearn.datasets import load_iris
iris = load_iris()
X, y = iris.data, iris.target

# Dividiendo los datos en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
#+Entrenar el modelo: Crear√°s una instancia de DecisionTreeClassifier y utilizar√°s el m√©todo fit para entrenar el modelo con los datos de entrenamiento.
clf = DecisionTreeClassifier()
clf.fit(X_train, y_train)
```

```python
#Hacer predicciones: Una vez que el modelo est√° entrenado, puedes utilizarlo para hacer predicciones con nuevos datos o con el conjunto de prueba.
y_pred = clf.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")
```

```python
#Visualizar el √°rbol: Scikit-learn tambi√©n ofrece herramientas para visualizar el √°rbol de decisi√≥n, lo que puede ayudarte a entender c√≥mo el modelo toma decisiones.
from sklearn import tree
import matplotlib.pyplot as plt

plt.figure(figsize=(12,12))
tree.plot_tree(clf, filled=True, feature_names=iris.feature_names, class_names=iris.target_names)
plt.show()
```
